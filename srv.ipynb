```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()


def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
```

```python
def get_topics_fome_file(fn):    
    l_perv = False
    temp_list = []
    temp_dict = dict()
    with open(fn) as fp:
        lines = fp.readlines()
        for line in lines:
            if '_' in line:
                l, r = line.split('_')
                l = l.strip()
    #             print(l)
    #             print(temp_list)
                if l.isnumeric() or l=='-1':
                    if l_perv:
                        temp_dict[int(l_perv)] = set(temp_list)
                    l_perv = l
                    temp_list = []
            else:
                line = line.lstrip()
                line = line.rstrip()
    #             print(line)
                if '*' not in line and line and 'Other' not in line:
                    temp_list.append(line)

    temp_dict[int(l)] = set(temp_list)
    return temp_dict        
```

```python
def count_dict_for_each_question(df, _related_words_by_topic, q=[]):
    num = 0
    q_count = 0
    q_portion = []
    f_col = ['Timestamp']
    count_dict = dict()
    for (column_name, column_data) in df.iteritems():
        col_values = []
        if column_name not in f_col:
            c_id, name = column_name.rsplit('-', 1)
            c_id = c_id.strip()
            c_id = int(c_id)

#             print(c_id)
#             print(ped_related_words_by_topic[c_id])
            for el in column_data:
                col_values.extend(list(set(el.split(';'))))
            ctr = Counter(col_values)
            
            o_flag = True
            for el in q:
                if el in ctr:
                    num = num + 1
                    q_count = q_count + ctr[el]
                    q_portion.append(ctr[el]/4)
                    o_flag = False
            if o_flag:
                q_portion.append(0)
                print(column_name)

#             print(ctr)
#             if 'pedestrian' in ctr:
#                 num = num + 1
#             elif 'Pedestrian' in ctr:
#                 num = num + 1
#             else:
#                 print(column_name)

            t_dict = dict()
            for el in _related_words_by_topic[c_id]:
                if el in ctr:
#                     print(el)
                    t_dict[el] = ctr[el]
                else:
                    t_dict[el] = 0
#             print(t_dict)
            count_dict[c_id] = t_dict
#             break
    print(num)
    print(q_count)
    return count_dict, q_portion
```
